```python
#This is a mini-game trial with use of random library of Python. 
#User is required to enter number of turns to play and an upper
#value for randint() method to work.

#Game provides feedback to the user in case which the user guess
#for a lower or a higher value than the generated number.

#Random Library Call
import random

#Methods for Inputs
def gettingUpperBound():
    inputvalue1 = int(input ("Enter an upper bound value as an integer:"))
    return inputvalue1

def gettingTurnNumber():
    inputvalue2 = int(input ("Enter rounds to play as an integer:"))
    return inputvalue2

def gettingGuess(iteration):
    inputvalue3 = int(input("Please enter your "+str(iteration)+". guess: "))
    return inputvalue3

##Welcome Banner
print("------------------------------------------------------------")
print("-------------Welcome to the Number Guess Game! -------------")
print("------------------------------------------------------------" + "\n")

print("The game generates a number in between 0 and a certain upper")
print("bound that you will set." +"\n")

##Exception handling
while (True):
    try:
        upperBound = gettingUpperBound()
    except: 
        print ("Try again. Please enter an Integer.")
    else:
        break

while (True):
    try:
        turnNumber = gettingTurnNumber()
    except: 
        print ("Try again. Please enter an Integer.")
    else:
        break

print("Guess a number in between 0 and "+str(upperBound) +"\n")
print("You will have "+ str(turnNumber) + " guesses!" +"\n")

print("------------------------Game Begins!------------------------" + "\n")

##Game mechanism
generatedNumber = random.randint(0,upperBound)

iteration = 1

while iteration <= turnNumber:
    
    while (True):
        try:
            playerGuess = gettingGuess(iteration)
        except: 
            print ("Try again. Please enter an Integer.")
        else:
            break

    guessedNumber = str(playerGuess)
    
    print("Round "+str(iteration))
    print("Your "+str(iteration)+". guess is: "+ guessedNumber)
    
    iteration = iteration+1
    if playerGuess == generatedNumber:
        print("You won!" + "\n")
        break
    if iteration == turnNumber+1:
        print("Out of Turns... You lost." + "\n")
        break
    if playerGuess > generatedNumber:
        print ("Please guess for a lower value." + "\n")
    else:
        print ("Please guess for a higher value." + "\n")

print("----------Thank you for playing Number Guess Game!----------")
```
